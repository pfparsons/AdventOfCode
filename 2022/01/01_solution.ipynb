```python
elves = [0]

csr = 0

with open("input", "r") as file:
    for line in file:
        if len(line) > 1:
            elves[csr] += int(line[:-1])
        else:
            elves.append(0)
            csr += 1
elves.sort(reverse=True)
elves[0]
            
```
